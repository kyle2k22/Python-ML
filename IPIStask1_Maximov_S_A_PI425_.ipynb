```python
import datetime
import numpy as np
import math
import matplotlib.pyplot as plt
dt = datetime.datetime.now()
plt.title(f'Вариант №{15}')
plt.ylabel('PI05/20b')
plt.xlabel('Maximov S.A.')
slx = []
gix = [0]
giy = [0.5]
enter = [1, -1, 3, 2]
exit = [1 , 1.5]
#Выполнил студент Максимов С.А. группы ПИ425
weights = np.array([[0.3, -0.5, 0.5, 0.1], [-0.8, -0.2, 0.2, 0.5]])
sum1 = sum(weights[0]*enter)
sum2 = sum(weights[1]*enter)
v1 = (math.exp(sum1)) / (1 + math.exp(sum1))
v2 = (math.exp(sum2)) / (1 + math.exp(sum2))
err = ((exit[0] - v1)**2 + (exit[1] - v2)**2) / 2
attempt = 0
print(f"Набор №{attempt}: V1: {'%.6f' % v1}, V2: {'%.6f' % v2}, Ошибка сети: {'%.6f' % err}")
while attempt < 3000:
  attempt +=1
  v1 = v1 * (1 - v1)
  v2 = v2 * (1 - v2)
  deltaves1 = err * v1
  deltaves2 = err * v2
  for i in range (len(weights)):
    for j in range (len(weights[i])):
      if j % 2 != 0:
        weights[i][j] += deltaves1 * (0.1 + 0.01)
      else:
        weights[i][j] += deltaves2 * (0.1 + 0.01)
  sum1 = sum(weights[0]*enter)
  sum2 = sum(weights[1]*enter)
  v1 = (math.exp(sum1)) / (1 + math.exp(sum1))
  v2 = (math.exp(sum2)) / (1 + math.exp(sum2))
  slx.append(err)
  err = ((exit[0] - v1)**2 + (exit[1] - v2)**2) / 2
  print(f"Набор №{attempt}: V1: {'%.6f' % v1}, V2: {'%.6f' % v2}, Ошибка сети: {'%.6f' % err}")
plt.plot(slx,color='r')
plt.show()
```
